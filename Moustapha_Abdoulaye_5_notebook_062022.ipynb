### III. Identification des outliers

#### III.1.  Representation graphique des valeurs aberrantes avec une boite à moustaches

```python
# Creation d'une boite à moustaches pour visualiser les valeurs aberrantes en utilisant seaborn
sns.boxplot(x=erp_liaison_web_product['price'])
```

- La boite à moustaches est un diagramme ou boxplot qui permet de disposer la dispersion et l'étalement des données. L'information est divisée en quatre parties appelées quartiles ou paramètres de disposition des données.
- La largeur du rectangle représente l'écart des interquartiles (quartile supérieur (Q3) - quartile inférieur (Q1))
- Les moustaches sont les deux lignes horizontales à l’extérieur de la boîte qui s’étendent du minimum jusqu’au quartile inférieur (le début de la boîte) et du quartile supérieur (la fin de la boîte) jusqu’au maximum.
- La moustache s’étire jusqu’à la valeur qui est la plus éloignée du centre, mais qui respecte une distance maximale de 1,5 fois l’écart interquartile en partant du quartile inférieur ou du quartile supérieur. Les données qui dépassent cette limite sont indiquées par des points et considérées comme potentiellement extrêmes.
- Les valeurs aberrantes sont les valeurs situées en dehors des moustaches inférieures et supérieures.

#### III.2.  Liste des valeurs aberrantes avec l'étendue interquartile (règle 1,5*IQR)

```python
# Calculons l'intervalle interquartile et le coefficient auxquels on appliquera les limites 
Q1 = erp_liaison_web_product['price'].quantile(0.25)
Q3 = erp_liaison_web_product['price'].quantile(0.75)
IQR = Q3 - Q1
coef = IQR * 1.5
```

- Le premier quartile, noté Q1, est la valeur au dessous de laquelle se trouvent 25% des données.
- Le troisième quartile, noté Q3, est la valeur au dessous de laquelle se trouvent 75% des données. 
- L'intervalle interquartile est la différence entre Q3 et Q1.
- Le coefficient met en évidence les valeurs aberrantes lorsqu'il est ajouté aux limites inférieures et supérieures.

```python
# fixons les limites pour voir à quelle distance les valeurs debordent des deux côtés
lim_inf = Q1 - coef
lim_sup = Q3 + coef
```

- Toutes valeurs superieures aux limites supérieures sont des valeurs aberrantes ou extrêmes
- Toutes valeurs inférieures aux limites inférieures sont des valeurs aberrantes ou extrêmes

```python
# cherchons les outliers
outliers = erp_liaison_web_product[(erp_liaison_web_product['price'] > lim_sup) | (erp_liaison_web_product['price'] < lim_inf)]
outliers
```

```python
# Trions les valeurs aberrantes des prix 
outliers.sort_values('price').head()
```

```python
# Trions les valeurs aberrantes des prix du plus élevé au moins élevé
outliers['price'].sort_values(ascending=False)
```

- Les prix qui s'écartent du reste des observations en forme de points sont nos valeurs aberrantes ou extrèmes. 
- Elles s'étalent de 85 € à 225 €.
